```
# prompt: cuantos paises tienen índices de co2 mayores a la media?

# Calculate the mean CO2 index
mean_co2 = df_nations["co2"].mean()

# Count the number of countries with CO2 index above the mean
num_countries_above_mean = len(df_nations[df_nations["co2"] > mean_co2])

print(f"Number of countries with CO2 indices above the mean: {num_countries_above_mean}")
```

```
#Usando where(, crearemos una variable nueva de co2, en donde identifiquemos qué países etan por debajo de la media, y qie paises por sobre la media.)
df_nations["co2_recodificada"] = np.where(df_nations["co2"] > df_nations["co2"].mean(), 1, 0)
```
